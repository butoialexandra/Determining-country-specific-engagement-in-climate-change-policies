```python
import pandas as pd
import glob
import re
```

```python
docs = [f for f in glob.glob("../txts/*.txt")]
```

```python
docs
```

```python
def country_name(doc):
    pos1 = doc.find('_')+1
    pos2 = doc.find('_', pos1)
    return doc[pos1:pos2]
```

```python
texts = []
countries = []
```

```python
for doc in docs:
    with open(doc, 'r') as f:
        text = f.read()
    texts.append(text)
    countries.append(country_name(doc))
```

```python
countries
```

```python
data = {
    'Country': countries,
    'NDC': texts
}
df = pd.DataFrame.from_dict(data)
```

```python
df
```

```python
df['NDC_processed'] = df['NDC'].apply(lambda x: re.sub('[(),\.!?0-9]', '', x)) # remove punctuation and numbers
```

```python
df['NDC_processed'] = df['NDC_processed'].apply(lambda x: x.lower())
```

```python
df['NDC_processed'] = df['NDC_processed'].apply(lambda x: x.strip())
```

```python
df['NDC_processed'] = df['NDC_processed'].apply(lambda x: re.sub(' +', ' ', x))
```
